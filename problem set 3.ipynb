```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
US_divorce = pd.read_csv('us-marriages-divorces.csv')
US_divorce
```

```python
US_divorce.head()
```

```python
years = US_divorce['Year']
marriages = US_divorce['Marriages_per_1000']
divorces = US_divorce['Divorces_per_1000']
US_divorce = plt.figure(figsize= (16,8))
US_divorce = plt.plot(years, marriages, label ='Marriages')
US_divorce = plt.plot(years, divorces, label = 'Divorces')
US_divorce = plt.title("marriages and divorces per capita in the U.S. (1867 to 2014)", fontsize=16)
US_divorce = plt.xlabel("Years", fontsize=16)
US_divorce = plt.legend(fontsize = 14)
US_divorce = plt.grid(True)
US_divorce
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
US_divorce = pd.read_csv('us-marriages-divorces.csv')
US_divorce
V = US_divorce.loc[US_divorce.Year.isin([1900,1950,2000]), ['Year', 'Marriages_per_1000', 'Divorces_per_1000']]
V
```

```python
V = V[V['Year'].apply(lambda x: x in [1900, 1950, 2000])]
years = V['Year']
marriages = V['Marriages_per_1000']
divorces = V['Divorces_per_1000']
V = plt.figure(figsize= (16,8))
V = plt.bar(years, marriages, label ='Marriages')
V = plt.bar(years, divorces, label = 'Divorces')
V = plt.title("Number of marriages and divorces per capita in the U.S. between 1990,1950 and 2000", fontsize=16)
V = plt.xlabel("Years", fontsize=14)
V = plt.legend(fontsize = 12, loc = "upper left")
V
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Scatter_plot = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
Scatter_plot
```

```python
years = Scatter_plot['Year']
arcade = Scatter_plot['Total Arcade Revenue (billions)']
computerAward = Scatter_plot['Computer Science Doctorates Awarded (US)']
Scatter_plot = plt.figure(figsize= (16,8))
Scatter_plot = plt.scatter(years, arcade, label ='Total Arcade Revenue (billions)')
Scatter_plot = plt.scatter(years, computerAward, label = 'Computer Science Doctorates Awarded (US)')
Scatter_plot = plt.title("Total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.", fontsize=16)
Scatter_plot = plt.xlabel("Years", fontsize=14)
Scatter_plot = plt.legend(fontsize = 12, loc = "upper left")
Scatter_plot = plt.grid(True)
Scatter_plot
```
